```python
import nltk
nltk.download('stopwords')
```

```python
stem=PorterStemmer()
```

#Creating a function for removing prefix and suffix. And contain only root word.

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
df1['content']=df1['content'].apply(stemming)
```

```python
print(df1['content'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X=df1['content'].values
Y=df1['label'].values
print(X)
print(Y)
```

#Converting words in digits or numbers

```python
vectorizer=TfidfVectorizer()
vectorizer.fit(X)
X=vectorizer.transform(X)
```

```python
print(X)
```

#Splitting into training and test data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=0,test_size=0.25,stratify=Y)
```

#Applying Logistic Regression model

```python
regressor=LogisticRegression()
regressor.fit(X_train,Y_train)
```

```python
X_train_prediction=regressor.predict(X_train)
```

```python
#Finding accuracy score for training data
from sklearn.metrics import accuracy_score
score=accuracy_score(Y_train,X_train_prediction)
print('Accuracy score of training model is:',score)
```

```python
#Finding accuracy score for test data
X_test_prediction=regressor.predict(X_test)
score2=accuracy_score(Y_test,X_test_prediction)
print("Accuracy score of test data is",score2)
```
